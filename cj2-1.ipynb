```python
import matplotlib.pyplot as plt

x = []
for i in range(101):
        x.append(float(0 + 0.01 * i))
        

def exp(x):
    temp = []
    for i in x:
        temp.append(i ** i)
    return temp
```

```python
y = exp(x)

plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig("Prompt1.pdf", format="pdf")
plt.show()
```
