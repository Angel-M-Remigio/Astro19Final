```python
import matplotlib.pyplot as plt

x = []
for i in range(101):
        x.append(float(0 + 0.01 * i))
```

```python
import math

def my_sin(x):
    sin = []
    for i in x:
        sin.append(math.sin(i))
        
    return sin

def my_cos(x):
    cos = []
    for i in x:
        cos.append(math.cos(i))
    
    return cos
```

```python
f, (ax1, ax2) = plt.subplots(1,2, sharey=False)
ax1.set_title("sin(x)")
ax1.plot(x, my_sin(x))

ax2.set_title("cos(x)")
ax2.plot(x, my_cos(x))

plt.savefig("Prompt2.pdf", format="pdf")
plt.show()
```
